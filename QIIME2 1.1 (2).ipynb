# Importación de lecturas

Las secuencias pareadas se importan a un artefacto de QIIME 2 (.qza) utilizando un archivo manifest que indica la ubicación de cada archivo FASTQ asociado a las muestras.

`SampleData[PairedEndSequencesWithQuality]`: datos de secuenciación paired-end con información de calidad.
`manifest.csv`: archivo que relaciona cada muestra con sus archivos R1 y R2.
`.qza`: formato de artefacto de QIIME 2.

```python
!qiime tools import \
  --type 'SampleData[PairedEndSequencesWithQuality]' \
  --input-path manifest.csv \
  --input-format PairedEndFastqManifestPhred33 \
  --output-path four-points.qza
```

# Evaluación inicial de calidad

Se genera una visualización (.qzv) que permite inspeccionar la calidad de las lecturas, el número de secuencias por muestra y otros parámetros importantes para decidir los pasos posteriores de filtrado y denoising.

```python
!qiime demux summarize \
  --i-data four-points.qza \
  --o-visualization four-points.qzv
```

# Filtrado de calidad QC y denoising con DADA2

Control de calidad y eliminación de ruido de los datos
Tener en cuenta el largo del amplicón, el fragmento de la secuencia y si es suficiente para hacer overlap entre las reads.

En esta versión de QIIME2 (2026.1), dada2 denoise-paired requiere 4 outputs obligatorios, no 3 como en versiones más viejas. El nuevo output que se agregó es --o-base-transition-stats, que genera una tabla con las tasas de transición de nucleótidos por score de calidad.

```python
!qiime dada2 denoise-paired \
  --i-demultiplexed-seqs four-points.qza \
  --p-trim-left-f 0 \
  --p-trunc-len-f 270 \
  --p-trim-left-r 0 \
  --p-trunc-len-r 270 \
  --o-representative-sequences repseq-fourpoints.qza \
  --o-table table-fourpoints.qza \
  --o-denoising-stats stats-dada2-fourpoints.qza \
  --o-base-transition-stats basestats-fourpoints.qza \
  --p-n-threads 0 --verbose
```

```python
!qiime metadata tabulate \
  --m-input-file stats-dada2-fourpoints.qza \
  --o-visualization stats-dada2-fourpoints.qzv
```

```python
!qiime feature-table summarize \
  --i-table table-fourpoints.qza \
  --m-metadata-file metadata-fourpoints.tsv \
  --o-feature-frequencies feature-freq-fourpoints.qza \
  --o-sample-frequencies sample-freq-fourpoints.qza \
  --o-summary table-fourpoints.qzv
```

```python
!qiime feature-table tabulate-seqs \
  --i-data repseq-fourpoints.qza \
  --o-visualization repseq-fourpoints.qzv
```

# Asignación taxonómica de secuencias 

La base de datos que elijas va a determinar en gran medida que tantos éxitos tengas en asignar un “nombre” a un taxón. Hay varias bases populares, sin embargo parece que SILVA es la más actualizada para el caso de bacterias (16s) y para el caso de hongos la base de datos UNITE (ITS), ambas son independientes de Qiime2.
Para asignar taxonómicamente necesitas un clasificador (classifier.qza), la manera más sencilla de obtener uno, es descargarlo de algún sitio. Qiime2 tiene una página de recursos donde se encuentran algunos, aunque, no están siempre actualizados. Para obtener la versión más nueva puedes entrenar tu clasificador de acuerdo a tus primers y secuencias objetivo siguiendo tutoriales dedicados a ello en la página de Qiime2.

```python
!qiime feature-classifier classify-sklearn \
  --i-classifier silva-138-v3v4-classifier.qza \
  --i-reads repseq-fourpoints.qza \
  --o-classification taxonomy-fourpoints.qza \
  --p-n-jobs 0
```

```python
!qiime metadata tabulate \
  --m-input-file taxonomy-fourpoints.qza \
  --o-visualization taxonomy-fourpoints.qzv
```

```python
!qiime taxa barplot \
  --i-table table-fourpoints.qza \
  --i-taxonomy taxonomy-fourpoints.qza \
  --m-metadata-file metadata-fourpoints.tsv \
  --o-visualization taxa-bar-plots.qzv
```
